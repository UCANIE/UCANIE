```python
import pandas as pd 

res = supabase.table('Exercise_duplicate').select("*").execute()
ori = pd.read_csv("Exercise.csv")
compa = pd.DataFrame(res.data)

print(ori.shape)
print(compa.shape)
```

```python
compa.head()
```

```python
# id를 기준으로 두 데이터프레임을 병합 (inner join)
merged_df = compa.merge(ori[['id', 'Name']], on='id', suffixes=('_compa', '_ori'))
# 병합된 데이터프레임의 필요한 부분 선택
update_data = merged_df[['id', 'Name_ori']].rename(columns={'Name_ori': 'Name'})
update_data.head()
```

```python

# Supabase upsert를 위해 데이터프레임을 딕셔너리 리스트로 변환
update_dict = update_data.to_dict(orient='records')
print(update_dict)
```

```python
# Supabase Exercise 테이블에 upsert
response = supabase.table('Exercise_duplicate').upsert(update_dict).execute()

# 결과 출력
print(response)
```

```python
import pandas as pd 

res = supabase.table('Exercise_duplicate').select("*").execute()
ori = pd.read_csv("Exercise.csv")
compa = pd.DataFrame(res.data)

print(ori.shape)
print(compa.shape)
```

```python
# id를 기준으로 두 데이터프레임을 병합 (inner join)
merged_df = compa.merge(ori[['id', 'Name', 'ClipSource', 'Multiply', 'Side']], on='id', suffixes=('_compa', '_ori'))

# 병합된 데이터프레임의 필요한 부분 선택
update_data = merged_df[['id', 'Name_ori', 'ClipSource_ori', 'Multiply_ori', 'Side_ori']].rename(
    columns={
        'Name_ori': 'Name',
        'ClipSource_ori': 'ClipSource',
        'Multiply_ori': 'Multiply',
        'Side_ori': 'Side'
    }
)

# Supabase update를 위해 데이터프레임을 딕셔너리 리스트로 변환
update_dict = update_data.to_dict(orient='records')
```

```python
# Supabase Exercise 테이블에 update
for record in update_dict:
    print(record)
    response_update = supabase.table('Exercise').update(record).eq('id', record['id']).execute()
    print(f"Update response for id {record['id']}:", response_update)
```
